## MatplotLib 

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Ot, or GTK+.

Some of the major Pros of Matplotlib are:

- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline 
# whenever you want to display any matplotlib graphs or any diagrams you always have to use plt and plot.show but using %matplotlib inline you do not have to use plot.show
```

```python
import numpy as np
```

```python
## Simple Examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
## plotting using matplotlib

##plt Scatter

plt.scatter(x,y,c='g') #g means green
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')
```

```python
y = x*x
```

```python
## plt plot

plt.plot(x,y,'r*--',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2d Diagram')
```

```python
## Creating Subplots
## withing one diagram you can create multiple subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'g*-')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Matplotlib demo')
plt.xlabel('x axis caption')
plt.ylabel('y axis caption')
plt.plot(x,y)
plt.show()
```

```python
## Compute the x and y coordinates for points on a sine curve

x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title('sine wave form')

# Plot the points using matplotlib
plt.plot(x, y)
plt.show()
```

```python
## Subplot()
## Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

## Set up a subplot grid that has height 2 and width 1,
## and set the first such subplot as active.
plt.subplot(2, 1, 1)

## Make the first plot
plt.plot(x, y_sin,'r--')
plt.title('Sine')

## Set the second subplot as active, and make the second plot.
plt.subplot(2, 1, 2)
plt.plot(x, y_cos,'g--')
plt.title('Cosine')
```

```python
## Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x, y, align = 'center')
plt.bar(x2, y2, color = 'g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

### Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

## using .random.narmal I am selecting a normal distribution btween some lower value to standard distribution value using a step size of 100
## rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);

## boxplot helps you to find percentiles
```

## Pie Chart

```python
## Data to Plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0) # explode 1st slide

## Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
```
